## Aproximación d EDPs de Poisson usando nubes de puntos
Resolver analíticamente una EDP puede llegar a ser imposible, por lo que se han desarrollado métodos núméricos para aproximar sus soluciones en una región. Tal región se delimitará por puntos, con los cuales se generarán contornos. Usaremos después un algoritmo de avance de frente para generar una nube de puntos dentro de la figura, y finalmente el algoritmo FPM para aproximar la EDP en tal región.

## Creación de la región

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Make grid
nx, ny = (5, 5) #points for x and y axis
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
xv, yv = np.meshgrid(x, y)
#store dots in df to apply conditions 
df = pd.DataFrame({"x":xv.flatten(), "y":yv.flatten()})
#df = df.loc[(abs(df["x"]) > 1 - 1e-16) | (abs(df["y"]) > 1 - 1e-16)]# use only corner dots
df = df.loc[(df["x"] == max(df["x"])) | (df["x"] == min(df["x"])) | (df["y"] == max(df["y"])) | (df["y"] == min(df["y"]))]# use only corner dots
#generate circle dots
def circle_points(r, n, center_x, center_y):
    circles = []
    t = np.linspace(0, 2*np.pi, n, endpoint=False)
    x = r * np.cos(t) + center_x
    y = r * np.sin(t) + center_y
    return x, y

radius = 0.2
n_dots = 10
x, y = circle_points(radius, n_dots, 0.5,0.5)
for xi, yi in zip(x, y):
    df.loc[df.index[-1]+1] = [xi, yi]
```

## Clase para aproximar la EDP

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import LineString
import copy
from numpy.linalg import eig
from sympy import *
from sympy.parsing.sympy_parser import parse_expr
#%matplotlib qt 

# settings #
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = [8, 8]
plt.figsize=(20, 20)
#fig, ax = plt.subplots()

class AdvancingFront():
    def __init__(self, contour_dots):
        ''' 
        This class fills a grid using an Advancing Front method
  
        Parameters:
        contour_dots (Dataframe): Dataframe containing "x" and "y" coordinates of contour dots
        self.contours (array): Stores arrays with ordered dots defining contours
        self.contour_type (array): Stores "internal" or "external" value for self.contours array
        '''
        self.dots = contour_dots.copy()
        self.dots["contour"] = True
        self.contours = None
        self.contour_type = None
        #Advancing front params
        self.r = None
        self.R = None
    def set_contours(self):
        '''
        This function makes relationships between a dot and the next dot in order to get sequences.
  
        Parameters:
        None
  
        Returns:
        Column "next" as the relation of dot and next one
        '''
        #set column for dot order relationship
        self.dots["next"] = ''
        curr_index = ''
        while True:
            #get dots that have an incoming connection
            used_incoming = self.dots["next"][ self.dots["next"] != '' ]
            #the available dots are the ones that are not from the above
            available_incoming = self.dots[~self.dots.index.isin(used_incoming)]
            #get dots that dont have an outgoing connection
            available_outgoing = self.dots[self.dots["next"] == '']
            if len(available_outgoing) == 0:
                break #finished...
            if curr_index == '':
                curr_index = available_outgoing.index[0]
            else:
                curr_incoming_dot = self.dots[ self.dots["next"] == curr_index ].index[0]
                if curr_incoming_dot in available_incoming.index: available_incoming = available_incoming.drop( curr_incoming_dot )
            points = available_incoming[["x","y"]].values
            single_point = self.dots.loc[curr_index, ["x","y"]].tolist()
            distances = self.euclid_dist(single_point, points)
            distances[distances < 1e-32] = 10000000 #dont consider the distance to same dot
            min_dist_index = np.argmin(distances)
            next_index = available_incoming.index[min_dist_index]
            self.dots.loc[curr_index, "next"] = next_index
            # if new dot has outgoing dot, then we have ended the contour...
            if self.dots.loc[next_index, "next"] != '':
                curr_index = ''
            else:    
                curr_index = next_index
        #display(self.dots)
    
    def draw_contours(self, draw_boundary_vectors = False, show_building = False, show_id = False):
        '''
        Draws contours using self.contours array and self.contour_type for colors.
        '''
        contours = self.get_contours()
        for i in range(self.get_contours_number()):
            contour = contours[i]
            for j in range(len(contour)):
                p1 = contour[j]
                p2 = contour[j+1] if j+1 < len(contour) else contour[0]
                plt.arrow(p1["x"],p1["y"], p2["x"]-p1["x"], p2["y"]-p1["y"], width = 0.007, 
                          color = "b" if self.get_contour_type(i) == "external" else "r")
                if show_id:
                    plt.annotate(p1.name, # this is the text
                     (p1["x"],p1["y"]), # these are the coordinates to position the label
                     textcoords="offset points", # how to position the text
                     xytext=(10,10), # distance from text to points (x,y)
                     ha='center') # horizontal alignment can be left, right or center
                if draw_boundary_vectors:
                    custom_size = 0.2
                    plt.arrow(p1["x"],p1["y"], p1["boundary_x"]*custom_size, p1["boundary_y"]*custom_size, 
                              width = 0.007, color = "gray" )
                if show_building: plt.pause(0.05)
                #plt.legend()
    def get_contours(self, columns = "all"):
        '''
        Separates dots indexes of contours in arrays.
        
        Returns an array of contours of dimensions (c, kc, m) with c the number of contours, kc the
        number of dots of the contour and all the columns m of the dataframe
        '''
        if self.contours is not None:
            return self.build_contours(self.contours, columns)
        
        contours = self.dots[self.dots["contour"] == True].copy()
        contours["used"] = False
        result = []
        while True:
            contour = []
            unused = contours[contours["used"] == False]
            if len(unused) == 0: break
            initial_index = unused.index[0]    
            p1 = unused.iloc[0]
            contour = contour + [unused.index[0]]
            contours.loc[ initial_index ,"used"] = True
            while True:
                p2 = unused.loc[p1["next"], :]
                contour = contour + [p1["next"]]
                contours.loc[ p1["next"] ,"used"] = True
                p1 = p2
                if p1["next"] == initial_index: break
            result = result + [contour]
        self.contours= result
        return self.build_contours(self.contours, columns)
    
    def build_contours(self, contours_indexes, columns = "all"):
        contours = copy.deepcopy(contours_indexes)
        #print(contours_indexes)
        for i in range(len(contours)):
            contour = contours_indexes[i]
            for j in range(len(contour)):
                if columns == "all":
                    contours[i][j] = self.dots.loc[contour[j],:]
                else:
                    contours[i][j] = self.dots.loc[contour[j],columns]
        return contours
    
    def get_contours_number(self):
        '''
        Returns contours quantity
        '''
        return len(self.contours)
    
    def get_contour_type(self, index = None):
        '''
        Returns contour type in same order of self.contours array.
        '''
        if self.contour_type is not None:
            return self.contour_type  if index == None else self.contour_type[index]
        
        min_index = None
        min_value = 1000000000
        for i in range(self.get_contours_number()):
            #get min values of each contour (doesnt matter if its "x" or "y" coordinate)
            value = np.min(self.get_contours(columns = ["x"])[i] )
            if value < min_value:
                min_value = value
                min_index = i
        self.contour_type = []
        for i in range(self.get_contours_number()):
            tyype = "external" if i == min_index else "internal"
            self.contour_type+= [tyype]
        
        return self.contour_type  if index == None else self.contour_type[index]
    
    
    def advancing_front(self, r, R, iterations, plot_result = True):
        '''
        Main advancing front algorithm.
        '''
        self.r = r 
        self.R = R
        #make n iterations
        for i in range(iterations):
            #iterate over each row
            n_rows = len(self.dots)
            for row in range(n_rows):
                dot = self.dots.iloc[row]
                x = dot["x"]
                y = dot["y"]
                x1, y1 = self.get_random_dot(x, y)
                if self.is_dot_inside_figure(x1,y1) and self.conditions_met(x1, y1):
                    self.dots.loc[self.dots.index[-1]+1] = [x1,y1, False] + ['']*(len(self.dots.columns)-3)
        if plot_result:
            self.draw_contours()
            dots = self.dots[self.dots["contour"] == False]
            plt.plot(dots.x, dots.y, marker='o', color='k', linestyle='none', markersize=2)
        plt.show()
    def is_dot_inside_figure(self, x, y):
        '''
        Checks if dot is inside figure using the ray test.
        
        info: https://blogs.20minutos.es/mati-una-profesora-muy-particular/tag/curvas-de-jordan/
        '''
        inside = True
        for contour, cntr_type in zip(self.get_contours(), self.get_contour_type()):
            cuts = self.get_ray_vs_contour_cuts(x,y, contour)
            if (cuts %2 == 0  and cntr_type == "external") or (cuts % 2 == 1 and cntr_type == "internal"):
                inside = False
                break
        return inside
    
    def get_boundary_vectors(self, draw = True, show_building = False, radius = 0.2, use_n_neigh = "all"):
        if "boundary_x" in self.dots.columns:
            self.draw_contours(draw_boundary_vectors = True, show_building = show_building)
            return #already executed algorithm
        
        self.dots["boundary_x"] = np.nan
        self.dots["boundary_y"] = np.nan
        dots = self.dots[self.dots["contour"] == True]
        
        for i in range(len(dots)):
            #dot = dots.iloc[i]
            #pdot = dots[ dots["next"] == dots.index[i] ].iloc[0] #prev dot
            #ndot = dots.loc[int(dot["next"]), :] #next dot
            #x,y = self.get_boundary_vector_end_dot(
            #    pdot["x"], pdot["y"], ndot["x"], ndot["y"], dot["x"], dot["y"] )
            x , y = self.find_outward_normal_vector(dots.index[i], radius = radius, use_n_neigh = use_n_neigh)
            self.dots.loc[dots.index[i], "boundary_x"] = x
            self.dots.loc[dots.index[i], "boundary_y"] = y
            
        if draw: self.draw_contours(draw_boundary_vectors = True, show_building= show_building)
    
    def find_outward_normal_vector(self, dot_id, radius = 0.2, use_n_neigh = "all"):
        test_size = 1e-4 #test to know if end of vector is inside figure
        neighbors = self.get_neighbors(dot_id, radius = radius, just_contours = True, keep_n = use_n_neigh )
        ri = np.array([self.dots.loc[dot_id, ["x", "y"]]], dtype=float).T #column vector
        C = np.array([[0,0],[0,0]])
        for index, rj in neighbors.iterrows():
            rj = np.array([rj[["x", "y"]]], dtype=float).T #column vector
            rji = rj-ri
            sq_norm = ( (rji.T).dot(rji) )
            wij = self.weight_function( dist = np.sqrt(sq_norm), alpha = 1, h =1 ) 
            C = C + ( (rji).dot( (rji).T ) )/sq_norm*wij
        val,vec=eig(C)
        lamb1, lamb2 = val
        #print(dot_id)
        if abs(lamb1-lamb2) < 1e-4:
            n1 = -sum(neighbors.x- ri[0][0])
            n2 = -sum(neighbors.y- ri[1][0])
            coords = np.array([ [n1,n2] ], dtype=float).T
        else:
            betha_index = np.argmin(val)
            n = np.array([vec[betha_index]]).T
            n2 = np.array([ [n[0][0], -n[1][0]] ]).T #reflection of y-axis to take other angles that have same cos^2 values
            if self.is_dot_inside_figure( (ri + n2*test_size)[0][0], (ri+n2*test_size)[1][0] ):
                coords = -n2 #coords where vector ends
            elif self.is_dot_inside_figure( (ri - n2*test_size)[0][0], (ri - n2*test_size)[1][0] ):
                coords = n2 #coords where vector ends
            elif self.is_dot_inside_figure( (ri + n*test_size)[0][0], (ri+n*test_size)[1][0] ):
                coords = -n #coords where vector ends
            else:
                coords = n #coords where vector ends
        coords = coords/ np.linalg.norm(coords) #unit vector 
        return coords.squeeze()
    
    def weight_function(self, dist, alpha = 1, h =1):
        return np.exp( -alpha*dist**2/h**2 )
        
    
    def get_neighbors(self, dot_id, radius = 0.2, just_contours = False, keep_n = "all"):
        center_dot = self.dots.loc[dot_id, :]
        x, y = center_dot[["x", "y"]]
        if just_contours:
            dots = self.dots[self.dots.contour == True]
        else:
            dots = self.dots
        neigh = dots[  (self.dots.x - x)**2+ (self.dots.y - y)**2 <= radius**2 ]
        neigh = neigh[ ~neigh.index.isin([dot_id])]
        if keep_n == "all": #keep all
            return neigh
        #keep just specified
        neigh = neigh.copy()
        neigh["distance"] = (neigh.x - x)**2+ (neigh.y - y)**2
        neigh = neigh.sort_values(by=['distance'], ascending = True)
        return neigh.head(keep_n)
    
    def get_approximation(self, poisson = None, dirichlet = None,neumann = None):
        '''
        Calculates the approx. of differential equation on the surface given in class with 
        advancing front applied. 
        
        Parameters:
        poisson (str): diff equation to solve
        dirichlet (string or dict): Contour dots with that condition, if string expression, 
            applies that to all contour, if dict, is of the form 
            {string1 : [dot11, dot12, ...], string2 : [dot21, dot22,...],...} and applies 
            each expression stringi for all their respective dots in list
        neumann (string or dict): Contour dots with that condition, if numeric, applies that to all contour,
            if dict, is of the form {string1 : [dot11, dot12, ...], string2 : [dot21, dot22,...],...} 
            and applies each expression stringi for all their respective dots in list
        
        dotij are the ids of each dot.
        
        Note that if dirichlet and neumann params are in dict form, they can be applied together to different
        dots.
        
        Returns:
        Creates column 'u(x,y)' on dots dataframe attribute with the aprroximations of diff. eq. on each dot.
        '''
        #initial values
        self.dots["u(x,y)"] = 0
        self.init_dirichlet_dots(dirichlet)
        x, y = symbols("x y")
        poisson = parse_expr(poisson)
        error = 1000000 
        while error > 1e-2:
            error = 0
            den_error = 0
            for index, row in self.dots.iterrows():
                if self.is_dot_dirichlet(row, dirichlet): continue #dont iterate dirichlet contours
                if self.is_dot_neumann(row, neumann):
                    boundary = np.array([row[["boundary_x", "boundary_y"]]], dtype=float).T #column boundary vector
                else:
                    boundary = None
                #get neighbors
                neighbors = self.get_neighbors(dot_id = index, radius = 0.8, keep_n = 6)
                #calculate W and M
                dot = np.array([row[["x", "y"]]], dtype=float).T #column vector
                poi = float(poisson.subs([(x,dot[0][0]), (y,dot[1][0])]) )#poisson evaluation on dot
                W = self.get_W(row, neighbors, poisson = poi, neumann = neumann)
                M = self.get_M(row, neighbors, poisson = poi, boundary = boundary)
                b = self.get_b(row, neighbors ,poisson = poi, neumann = neumann)
                MTWM = M.T.dot(W).dot(M)
                a = np.linalg.inv(MTWM).dot(M.T).dot(W).dot(b)
                f = a[0][0]
                old_f= self.dots.loc[index, "u(x,y)"]
                self.dots.loc[index, "u(x,y)"] = f
                error += abs(f-old_f)
                den_error += abs(f)
                
            error = error/den_error
    def draw_approximation(self, as_surface = False):
        if as_surface:
            fig = plt.figure()
            ax = fig.gca(projection='3d')
            X,Y,Z = self.dots["x"].values, self.dots["y"].values, self.dots["u(x,y)"].values
            ax.plot_trisurf(X,Y,Z, linewidth=0.1, antialiased=True, cmap = "gray")
        else:
            ax = plt.axes(projection='3d')
            ax.scatter3D(self.dots["x"].values, self.dots["y"].values, self.dots["u(x,y)"].values, c='black')
            
            
    #HELP METHODS
    
    def init_dirichlet_dots(self, dirichlet):
        if dirichlet is None: return # no dirichlet condition
        #set dirichlet boudary values
        x, y = symbols("x y")
        if type(dirichlet) is dict:
            for value, dots in dirichlet.items():
                expr = parse_expr(value)
                for dot_id in dots:
                    dot = self.dots.loc[dot_id]
                    val = float(expr.subs([(x,dot["x"]), (y,dot["y"])]) )
                    self.dots.loc[dot_id,"u(x,y)"] = val
        else: #dirichlet is expression applied to all contour dots
            expr = parse_expr(dirichlet)
            dots = self.dots.loc[self.dots.contour == True]
            for index, dot in dots.iterrows():
                val = float(expr.subs([(x,dot["x"]), (y,dot["y"])]) )
                self.dots.loc[index,"u(x,y)"] = val
    
    def is_dot_dirichlet(self, row, dirichlet):
        if row["contour"] == False: return False
        if dirichlet is None: return False
        if type(dirichlet) is dict:
            for value, dots in dirichlet.items():
                if row.name in dots: return True
            return False #not in dictionary
        else: #dirichlet is number applied to all contour dots
            return True
    
    def is_dot_neumann(self, row, neumann):
        if row["contour"] == False: return False
        if neumann is None: return False
        if type(neumann) is dict:
            for value, dots in neumann.items():
                if row.name in dots: return True
            return False #not in dictionary
        else: #neumann is number applied to all contour dots
            return True
    
    def get_dot_neumann(self, row, neumann):
        if row["contour"] == False: return ''
        x, y = symbols("x y")
        if type(neumann) is dict:
            for value, dots in neumann.items():
                if row.name in dots:
                    expr = parse_expr(value)
                    val = float(expr.subs([(x,row["x"]), (y,row["y"])]) )
                    return val
            return '' #not in dictionary
        else: #neumann is number applied to all contour dots
            expr = parse_expr(neumann)
            val = float(expr.subs([(x,row["x"]), (y,row["y"])]) )
            return val
    
    def get_W(self, row, neighbors, poisson = None, neumann = None):
        '''Makes diagonal matrix W of size n with n the number of neighbors'''
        W_diag = np.array([])
        dot = np.array([row[["x", "y"]]], dtype=float).T #column vector
        for n_index, neigh in neighbors.iterrows():
            neigh = np.array([neigh[["x", "y"]]], dtype=float).T
            vec = dot - neigh
            sq_norm = vec.T.dot(vec)
            #calculate weight for each one
            W_diag = np.append(W_diag, self.weight_function(dist = np.sqrt(sq_norm), alpha = 1, h = 1) )
        if poisson is not None:
            W_diag = np.append(W_diag, 1)
        if self.is_dot_neumann(row, neumann):
            W_diag = np.append(W_diag, 1 )
        W = np.diag(W_diag)
        return W
    
    def get_M(self, row, neighbors, poisson = None, boundary = None):
        dot = np.array([row[["x", "y"]]], dtype=float).T #column vector
        x1 = dot[0][0]
        x2 = dot[1][0]
        for n_index, neigh in neighbors.iterrows():
            neigh = np.array([neigh[["x", "y"]]], dtype=float).T
            xi1 = neigh[0][0]
            xi2 = neigh[1][0]
            dxi = xi1 - x1
            dyi = xi2 - x2 
            if 'M' in dir():
                M = np.concatenate((M, [[1, dxi, dyi, 1/2*dxi**2, dxi*dyi, 1/2*dyi**2]]))
            else:
                M = np.array([[1, dxi, dyi, 1/2*dxi**2, dxi*dyi, 1/2*dyi**2]])     
        if poisson is not None:
            M = self.add_dims_to_matrix(M, add_row = True)
            M[M.shape[0]-1, [3, 5]]=1
        if boundary is not None:
            M = self.add_dims_to_matrix(M, add_row = True)
            M[M.shape[0]-1, [1, 2]]= boundary.flatten()
        return M
    
    def get_b(self, row, neighbors ,poisson = None, neumann = None):
        b = np.array([neighbors["u(x,y)"].values]).T
        if poisson is not None :
            b = np.concatenate((b, [[poisson]]), axis = 0)
        if self.is_dot_neumann(row, neumann):
            b = np.concatenate((b, [[self.get_dot_neumann(row, neumann)]]), axis = 0)
        return b
    
    def add_dims_to_matrix(self, mx, add_row = False, add_column = False):
        if add_row:
            mx = np.concatenate((mx, [[0]*mx.shape[1]]), axis = 0)
        elif add_column:
            mx = np.concatenate((mx, [[0]]*mx.shape[0]), axis = 1)
        else:
            mx = np.concatenate((mx, [[0]]*mx.shape[0]), axis = 1)
            mx = np.concatenate((mx, [[0]*mx.shape[1]]), axis = 0)
        return mx
        
    
    def get_ray_vs_contour_cuts(self, x,y, contour):
        '''
        Gets times a ray starting from (x,y) cuts a contour, cutting first the center of its first segment.
        '''
        cuts = 1 #cut of the first point where ray is directed
        dist = 100000
         #selected consecutive contour dots to make the first cut on the center of the segment generated
        dot2 = contour[0]
        dot3 = contour[-1]
        x2, y2 = (dot2["x"]+dot3["x"])/2, (dot2["y"]+dot3["y"])/2
        #This formula is used to generate a far dot to simulate a ray: y =m(x-x1)+y1
        if abs(x2-x) < 1e-32:
            ray = LineString([(x, y), (x,  y + 1000*(y<y2)   )])
        else:
            m = (y2-y)/(x2-x)
            if x2 < x: #negative if contour dot is to the left side of x
                dist = -dist
            ray = LineString([(x, y), (dist,  m*(dist-x)+y  )])
        
        for i in range(0, len(contour)-1): #dont use the dot used for first intersection
            p1, p2 = contour[i], contour[i+1]
            segment = LineString([(p1["x"], p1["y"]), (p2["x"], p2["y"])])
            if ray.intersects(segment):
                cuts+=1    
        return cuts
    
    def euclid_dist(self, single_point, points):
        '''
        Calculates euclidean distance of a single dot vs an array of dots
        
        Parameters:
        single_point (np array of dim (2,) ): Coordinates x and y of dot 
        points (np array of dim (k,2) ): Coordinates x and y of k dots 
  
        Returns:
        np array of dim (k,2) containing the euclidean distance of single point vs each dot of points array
        '''
        return np.sqrt(((single_point-points)**2).sum(axis = 1))
    
    
    
    def get_random_dot(self, x, y):
        '''
        Gets new random dot using another as a reference.
        '''
        r = self.r
        R= self.R
        #get a random dot bewteen short and long circunference
        angle = np.random.uniform(0, 2*np.pi)
        rand = np.random.uniform(0, 1)
        distance = r + (R-r)*rand #distance between r and R
    
        x1 = x + distance * np.sin(angle)
        y1 = y + distance * np.cos(angle)
        return x1, y1
    
    def conditions_met(self,x, y):
        #check if dot satisfy all conditions
        r = self.r
        #check if dot is outside all the circunferences of radius r
        for _, row in self.dots.iterrows():
            xi = row["x"]
            yi = row["y"]
            out = (x-xi)**2+(y-yi)**2 > r**2
            if not out: return False
        return True
    
    def get_boundary_vector_end_dot(self, x1, y1, x2, y2, x3, y3, alg = "PerpBis"):
        '''
        Gets the other dot of boundary vector (the one that gives orientation) using some technique.
        
        Techniques:
        
        Perpendicular bisector: Gets perpendicular bisector of (x1,y1)-(x2,y2) that goes through (x3,y3).
        
        Bisector ponderarion: Gets perpendicular bisector of (x1,y1)-(x3,y3) and (x2,y2)-(x3,y3)
                             and makes a ponderation using each segmnent length 
        
        Returns:
        Returns "x" and "y" coordinates of second dot of vector
        '''
        
        
        size = 1/10 #size of return
        test_size = 1/20 #size used to test if dot of vector is inside figure, make it smaller if gives problems
        m = (y1-y2)/(x1-x2)
        if abs(m) < 1e-32: #m is zero
            x = x3
            y = (y3+1)
        else:
            x = x3+1
            y = (-1/m)*(x-x3)+y3
        unit_x = (x-x3)/((x-x3)**2+(y-y3)**2)**(1/2)
        unit_y = (y-y3)/((x-x3)**2+(y-y3)**2)**(1/2)
        
        x = x3 + unit_x*test_size
        y = y3 + unit_y*test_size
        if self.is_dot_inside_figure(x,y): #change orientation of vector if (x,y) is inside figure
            x = x3 - unit_x*size
            y = y3 - unit_y*size
        else:
            x = x3 + unit_x*size
            y = y3 + unit_y*size  
        
        return x, y
```

# Demostración de algoritmo para obtener vectores normales a los contornos

Se discute la sección 4 del paper normales_oprimizacion.pdf

Dado que $C$ es la suma de matrices simétricas, $C$ es una matriz simética, lo que equivale a ser una matriz ortogonalmente diagonalizable, es decir, $C= PDP^T$, con $P$ una matriz ortogonal (formada por los eigenvectores normales como columnas) y $D$ una matriz diagonal (formada por los eigenvalores respectivos). De allí,

$$\sum_j{\cos^2{\alpha_{ij}w_{ij}}} = \textbf{n}^TC\textbf{n} = (\textbf{n}^TP)D(P^T\textbf{n})=(P^T\textbf{n})^TD(P^T\textbf{n})= \textbf{y}^TD\textbf{y}= \sum_{i=1}^n{\lambda_iy_i^2}$$

Con $\lambda_i$ los eigenvalores de $C$ y $y_i$ los elementos del vector $\textbf{y}= P^T\textbf{n}$. Es importante mencionar que $||\textbf{y}||=1$, pues, como $P$ es una matriz ortogonal, $P^T$ también, y en consecuencia, se tiene que $ ||P^T\textbf{x}||=||\textbf{x}||, \hspace{2mm} \forall \textbf{x}$, es decir, la norma se preserva en el producto de un vector por una matriz ortogonal.

Encontrar el mínimo valor es equivalente entonces a lo siguiente:

$$\min\left\{\textbf{n}^TC\textbf{n}, \hspace{2mm} ||\textbf{n}|| = 1 \right\} = \min\left\{\sum_{i=1}^n{\lambda_iy_i^2}, \hspace{2mm} ||\textbf{y}|| = 1\right\} $$

Si queremos minimizar esa suma, es lógico poner toda la magnitud del vector en el eigenvalor de menor valor. Es decir, **en caso de que el eigenvalor mínimo tenga multiplicidad algebraica uno** (único en la expresión), entonces esa i-ésima componente cuadrada vale uno y las demás cero, es decir, al tomar raíces es un vector canónico $\textbf{e}_i$ (o su inverso aditivo), con $i$ el índice del eigenvalor menor. Es más, esos dos vectores serán los únicos que minimizan. En resumen, tendremos,

$$\textbf{y}=\textbf{e}_i$$

$$\textbf{y}=-\textbf{e}_i$$

Y todavía debemos obtener $\textbf{n}$:

$$\textbf{n}= P\textbf{y}=P\textbf{e}_i=\textbf{p}_i=\textbf{v}_i$$

$$\textbf{n}= P\textbf{y}=P(-\textbf{e}_i)=-\textbf{p}_i=-\textbf{v}_i$$

con $\textbf{v}_i$ el eigenvector asociado a $\lambda_i$ por construcción de $P$. Debemos recordar que la norma se preserva, por lo que $\textbf{n}$ ya es normal.

En resumen, se debe hacer lo siguiente **en caso de que el eigenvalor mínimo tenga multiplicidad algebraica uno**:

1. Encontramos eigenvalores y eigenvectores de $C$
2. Buscamos el eigenvalor mínimo, el cual debe ser único
3. El vector $\textbf{n}$ buscado es el eigenvector normal asociado, si no está fuera de la figura, usamos $-\textbf{n}$ 

En caso de que **el eigenvalor mínimo tenga multiplicidad algebraica mayor a uno**, se tendrían infinitas soluciones, pues hay infinitas combinaciones lineales de los mismos con coeficientes que suman uno.

Por ejemplo tomemos un caso en dimensión 2, si $C = aI$, sus eigenvalores asociados son $\lambda_1 = a = \lambda_2$, es decir, formalmente es uno con multiplicidad algebráica dos. Además, podemos notar que cualquier vector normal da el mismo resultado en la expresión a minimizar:

$$\textbf{n}^TC\textbf{n}=\textbf{n}^T(aI)\textbf{n}= a \textbf{n}^TI\textbf{n}= a\textbf{n}^T\textbf{n}= a||\textbf{n}||^2=a$$

Es decir, cualquier vector normal es mínimo. El algoritmo anterior regresaría alguno de los cuatro eigenvectores normales

$$(1,0), (0,1), (-1,0), (0,-1)$$

Lo cual puede no ser la respuesta buscada.
