# Ajuste por curva de Richards

Una curva de Richards es la solución de la ecuación diferencial

$$\frac{\partial P}{\partial t}=rP(1-\frac{P^\beta}{k^\beta}), \hspace{2mm}  \beta \in \mathbb{R}$$

Y es de la siguiente forma:

$$P(t)=\frac{P_0k}{(  P_0^\beta+(k^\beta-P_0^\beta)e^{-\beta r(t-t_0)} )^{1/\beta}}, \hspace{2mm} P(t_0)=P_0$$

Ya que el método de Broyden necesita estar "cerca" de una solución para moverse hacia ella, se graficaron los puntos escalados en geogebra y se ajustó una "buena" curva a éstos usando deslizadores para los parámetros involucrados.

*Nota*: Para los valores del eje $x$ se comenzó desde 0 y se aumentó de 1 en 1 para cada dato, al escalar se dividió entre el valor de $x$ más grande. Para los valores del eje $y$ se tomaron los datos, y se dividió entre el dato mayor de los acumulados.

![Curva aproximada por geogebra](muertes_richards.png)

El método de Broyden que ejecutaremos trabaja sobre los datos escalados, es decir, los valores de $r,k,\beta$ que obtiene no son una solución óptima del problema real.

Sean $\delta_t$ el dato mayor en tiempo y $\delta_p$ el dato mayor en población con los cuales se escala la información, y $r,k,\beta$ las soluciones obtenidas del método de Broyden con datos escalados, entonces

$$ k_{real} = k \delta_p $$

$$ r_{real} = \frac{r}{\delta_t} $$

$$ \beta_{real} = \beta $$

**Nota**: En el método de Broyden escribimos a $\beta$ como $b$.

Para los puntos iniciales del algoritmo, se construye una cuadrícula cerca de la solución aproximada en geogebra, es decir, cerca del punto $(r=330,k=1,b=0.027)$. Además, de matriz inicial se usa $DF^{-1}$ para aumentar la probabilidad de que llegue a una solución


```python
#Este programa encuentra los parámetros óptimos r y k para ajustar 
#una curva de Richards
#Función que calcula el error: E(r,k) = sigma_i=1^n(  P(ti)-Pi  ), con Pi las mediciones dadas 
#Este algoritmo usa el método de Broyden con la variante de Sherman Morrison
from sympy import *
import matplotlib.pyplot as plt
from itertools import product

#definimos la funcion para modelar
def funcion_poblacional(t0,p0,k,r,t,b):
    #devuelve la función a usar uando los tipos y valores actuales de las variables
    #logistica (b=1): (k*p0)/(p0+(k-p0)*exp(-r*t))
    #r(P) cuadratica (b=2): P = p0*k/(  p0**2+(k**2-p0**2)*exp(-2*r*t) )**(1/2)
    #etc...
    #richards: p0*k/(  p0**b+(k**b-p0**b)*exp(-b*r*t) )**(1/b)
    return p0*k/(  p0**b+(k**b-p0**b)*exp(-b*r*(t-t0)) )**(1/b)

#definimos el sistema F para el cual queremos encontrar las raíces, i.e., donde F= vector 0
def F(coord):
    global fr, fk, fb
    #el vector vertical tiene una dimensión más, es de la forma [ [r], [k]]
    r1 = coord[0,0]
    k1 = coord[1,0]
    b1 = coord[2,0]
    return np.matrix([ fr.subs({r:r1, k: k1, b:b1 }), fk.subs({r:r1, k: k1,b:b1}), fb.subs({r:r1, k: k1, b:b1}) ]).T
def DF(coord):
    global frr, frk, frb, fkr, fkk, fkb, fbr, fbk, fbb
    #el vector vertical tiene una dimensión más, es de la forma [ [r], [k]]
    r1 = coord[0,0]
    k1 = coord[1,0]
    b1 = coord[2,0]
    
    return np.matrix([
        [frr.subs({r:r1, k: k1, b:b1 }), frk.subs({r:r1, k: k1,b:b1}), frb.subs({r:r1, k: k1, b:b1})],
        [fkr.subs({r:r1, k: k1, b:b1 }), fkk.subs({r:r1, k: k1,b:b1}), fkb.subs({r:r1, k: k1, b:b1})],
        [fbr.subs({r:r1, k: k1, b:b1 }), fbk.subs({r:r1, k: k1,b:b1}), fbb.subs({r:r1, k: k1, b:b1})]
    ]).astype(float)


def Broyden(punto_inicial, useDF = False):
    print("punto inicial")
    print(punto_inicial)
    
    #---- Método de Newton de varias variables (Broyden) ------
    Nmax = 50; #numero máximo de iteraciones del método
    tol = 10**(-4); #tolerancia relativa del método
    pt=np.matrix(punto_inicial).T; # punto inicial, converge con (5,1)
    for i in range(0,Nmax):
        if i == 0:
            if useDF:
                df = DF(pt)
                Ainv = np.linalg.inv(df)
            else:
                Ainv = np.matrix('1 0 0; 0 1 0; 0 0 1')
        else:
            yk = F(pt1) - Fx
            sk=pt1-pt
            prod_ext = (sk-Ainv.dot(yk) ).dot(sk.T)
            numerador = prod_ext.dot(Ainv)
            denominador = sk.T.dot(Ainv).dot(yk)
            Ainv = Ainv + (1/denominador[0,0])*numerador
            pt=pt1;    
        Fx = F(pt)
        pt1= pt-Ainv.dot(Fx)
        if np.linalg.norm(pt1.astype(float) - pt.astype(float))/np.linalg.norm(pt1.astype(float)) < tol and i > 5:
            print("Tolerancia relativa alcanzada en ", str(i+1), " iterciones" )
            break 
        if i == Nmax-1:
            print("Número máximo de iteraciones alcanzado")
    pt = pt1
    #----------- Datos Obtenidos ---------------
    #print("Candidato a Raiz")
    #print(pt)
    #print("Evaluacion en f1 y f2");
    #print(F(pt,fr,fk))
    print("r, k y b obtenidos en escala")
    r = pt[0,0]
    k = pt[1,0]
    b = pt[2,0]
    print(r)
    print(k)
    print(b)
    #print("r y k reales")
    #print(r/max_ti)
    #print(k*max_p)

    #ya obtenidos r y k, volvemos a la escala real
    #ti = ti*max_ti;
    #p = p*max_p;

    #obtenemos P
    t0 = ti[0]
    p0 = p[0]
    P = funcion_poblacional(t0,p0,k,r,t,b)
    print("Evaluación en la función error")
    error = 0
    for i in range(0,n):
        x = ti[i];
        error = error + (P.subs(t,x) - p[i] )**2;
    print(error)
    #------------------------Gráfica ------------------------
    fig, ax = plt.subplots() #inicializamos grafico
    #intervalo a graficar
    desde_x = min(ti) 
    hasta_x = max(ti)
    #graficamos puntos dados
    ax.scatter(ti, p)
    #Graficamos la funcion
    xs = np.linspace(desde_x, hasta_x, 50) #intervalo y cantidad de puntos a evaluar
    ys = [P.subs(t,k) for k in xs] #evaluaciones en la funcion encontrada
    ax.plot(xs, ys, 'r-', label="P(t)") #graficamos con una recta
    #Estilizado de gráfica
    plt.xlabel("Tiempo") #etiqueta para eje x
    plt.ylabel("Población") #etiqueta para eje y
    plt.legend(loc='best') #habilitar la muestra de etiquetas
    plt.show()
    #limpiar gráfico para la siguiente iteración
    plt.figure().clear()
    plt.close()
    plt.cla()
    plt.clf()

########## ALGORITMO PRINCIPAL ###########
    
# Datos del problema
ti = np.arange(0,datos.shape[0],1) # x = 0,1,2,..., (numero de filas-1).
#ti = np.array(range(0,datos.shape[0]))
p = np.array(datos["M. Totales"])          
n = ti.shape[0] #cantidad de datos

#escalar la información, así no se desbordan las operaciones
max_ti = max(ti)
max_p = max(p)

ti= ti/max_ti
p= p/max_p

t0 = ti[0]
p0 = p[0] #primer dato poblacional (escalado)
#declaracion de símbols
t, r, k, b  = symbols('t r k b', real=True)
#funcion poblacional con variables simbólicas
P = funcion_poblacional(t0,p0,k,r,t,b)
#funcion error
E = sum( (P.subs(t,ti[i])-p[i])**2 for i in range(0,n))
print("Calculando derivadas parciales para armar F")
#derivadas parciales (sistema F)
fr = diff(E,r)
fk = diff(E,k)
fb = diff(E,b)
print("Calculando segundas derivadas parciales para armar DF")
#segundas derivadas parciales (matriz DF)
frr = diff(fr,r)
frk = diff(fr,k)
frb = diff(fr,b)
fkr = diff(fk,r)
fkk = diff(fk,k)
fkb = diff(fk,b)
fbr = diff(fb,r)
fbk = diff(fb,k)
fbb = diff(fb,b)    

#puntos iniciales    
rs = np.arange(330, 331,1)
ks = np.arange(0.9, 1.01, 0.1)
bs = np.arange(0.025, 0.031, 0.005)
puntos_iniciales = list(product(rs,ks,bs))
#puntos_iniciales = [(330, 1.0, 0.03), (330, 1.2000000000000002, 0.015), (330, 1.2000000000000002, 0.03)]
print("Iniciamos método de Broyden")
for punto_inicial in puntos_iniciales:
    try:
        Broyden(punto_inicial, useDF = True)
    except TypeError:
        print("Cannot convert complex to float. Siguiente punto...")
    #except LinAlgError:
    #    print("Matriz inicial DF no invertible. Siguiente punto...")
```

## Confirmar que los puntos hallados son mínimos


Teorema 9.6(CALCULUS volumen 2. Tom M.Apostol): (https://es.wikipedia.org/wiki/Matriz_hessiana)

Sea $f$ un campo escalar con derivadas parciales segundas continuas en una bola } $B(\textbf{a})$, designemos como $H(\textbf{a})$ a la matriz hessiana en el punto estacionario $\textbf{a}$. Entonces:

$f$ tiene un mínimo relativo en $\textbf{a}$ si todos los autovalores $H(\textbf{a})$ son positivos ssi los determinantes de las submatrices principales de $H(\textbf{a})$ son positivos (teorema de Análisis 1, 2 de Marzo del 2022)

Es decir, calcularemos los determinantes de las submatrices principales. Si todos son positivos, tenemos un mínimo relativo.

```python
def funcion_poblacional(t0,p0,k,r,t,b):
    #devuelve la función a usar uando los tipos y valores actuales de las variables
    #logistica (b=1): (k*p0)/(p0+(k-p0)*exp(-r*t))
    #r(P) cuadratica (b=2): P = p0*k/(  p0**2+(k**2-p0**2)*exp(-2*r*t) )**(1/2)
    #etc...
    #richards: p0*k/(  p0**b+(k**b-p0**b)*exp(-b*r*t) )**(1/b)
    return p0*k/(  p0**b+(k**b-p0**b)*exp(-b*r*(t-t0)) )**(1/b)
# Datos del problema
ti = np.arange(0,datos.shape[0],1) # x = 0,1,2,..., (numero de filas-1).
#ti = np.array(range(0,datos.shape[0]))
p = np.array(datos["M. Totales"])          
n = ti.shape[0] #cantidad de datos

#escalar la información, así no se desbordan las operaciones
max_ti = max(ti)
max_p = max(p)

ti= ti/max_ti
p= p/max_p

t0 = ti[0]
p0 = p[0] #primer dato poblacional (escalado)
#declaracion de símbols
t, r, k, b  = symbols('t r k b', real=True)
#funcion poblacional con variables simbólicas
P = funcion_poblacional(t0,p0,k,r,t,b)
#funcion error
E = sum( (P.subs(t,ti[i])-p[i])**2 for i in range(0,n))
print("Calculando derivadas parciales para armar F")
#derivadas parciales (sistema F)
fr = diff(E,r)
fk = diff(E,k)
fb = diff(E,b)
print("Calculando segundas derivadas parciales para armar H")
#segundas derivadas parciales (matriz DF)
frr = diff(fr,r)
frk = diff(fr,k)
frb = diff(fr,b)
fkr = diff(fk,r)
fkk = diff(fk,k)
fkb = diff(fk,b)
fbr = diff(fb,r)
fbk = diff(fb,k)
fbb = diff(fb,b)

#puntos a evaluar
candidatos = [
    (471.330284588446, 0.945177586281642, 0.0199428740611559),
    (519.012200827836, 0.945338228744005, 0.0174561330444561),
    (333.143704519160, 0.940188492804259, 0.0288798589317606)
    ]
    
for candidato in candidatos:
    print ("Punto")
    print(candidato)
    r1 = candidato[0]
    k1 = candidato[1]
    b1 = candidato[2]
    H = np.matrix([
        [frr.subs({r:r1, k: k1, b:b1 }), frk.subs({r:r1, k: k1,b:b1}), frb.subs({r:r1, k: k1, b:b1})],
        [fkr.subs({r:r1, k: k1, b:b1 }), fkk.subs({r:r1, k: k1,b:b1}), fkb.subs({r:r1, k: k1, b:b1})],
        [fbr.subs({r:r1, k: k1, b:b1 }), fbk.subs({r:r1, k: k1,b:b1}), fbb.subs({r:r1, k: k1, b:b1})]
    ]).astype(float)
    print("Determinantes de las submatrices principales")
    #submatriz 1
    ixgrid = np.ix_([True, True, True], [True, True, True])
    H1 = H[ixgrid]
    print(np.linalg.det(H1))
    #submatriz 2
    ixgrid = np.ix_([False, True, True], [False, True, True])
    H2 = H[ixgrid]
    print(np.linalg.det(H2))
    #submatriz 3
    ixgrid = np.ix_([False, False, True], [False, False, True])
    H3 = H[ixgrid]
    print(np.linalg.det(H3))
```

Debido a que no estamos exactamente en un punto mínimo local en cada caso, y que la computadora no hace cálculos exactos, basta con ver que los tres determinantes sean más "cercanos" a ser positivos o negativos. En todos los casos puede verse que los detemrinantes tienden a ser positivos, es decir, son puntos cercanos a sus respectivos **mínimos relativos**.

# Ajuste por curva logística generalizada

Para resolver esta ecuación se tomará en cuenta una hipótesis adicional

$$\frac{\partial P}{\partial t}=rP^\alpha(1-\frac{P^\beta}{k^\beta}), \hspace{2mm} \alpha, \beta\in \mathbb{R}, \hspace{2mm} \alpha + \beta = 1$$

Su solución es de la forma:

$$P(t) = (k^\beta + (p_0^\beta - k^\beta)e^{-\beta \frac{r}{k^\beta}(t-t_0)})^{\frac{1}{\beta}}$$

Hacemos lo mismo queen el otro caso, de hecho, comenzamos en el mismo punto, pues la curva es muy parecida a la otra con esos parámetros, lo cual tiene sentido, pues en el caso anterior las soluciones cumplían que $\alpha + \beta \approx 1$ .

![Curva aproximada por geogebra](muertes_generalizada.png)

```python
#Este programa encuentra los parámetros óptimos r y k para ajustar 
#una curva logística generalizada
#Función que calcula el error: E(r,k) = sigma_i=1^n(  P(ti)-Pi  ), con Pi las mediciones dadas 
#Este algoritmo usa el método de Broyden con la variante de Sherman Morrison
from sympy import *
import matplotlib.pyplot as plt
from itertools import product

#definimos la funcion para modelar
def funcion_poblacional(t0,p0,k,r,t,b):
    #devuelve la función a usar uando los tipos y valores actuales de las variables
    return (k**b + (p0**b - k**b)*exp(-b*(r / k**b)*(t-t0)))**(1 / b)

#definimos el sistema F para el cual queremos encontrar las raíces, i.e., donde F= vector 0
def F(coord):
    global fr, fk, fb
    #el vector vertical tiene una dimensión más, es de la forma [ [r], [k]]
    r1 = coord[0,0]
    k1 = coord[1,0]
    b1 = coord[2,0]
    return np.matrix([ fr.subs({r:r1, k: k1, b:b1 }), fk.subs({r:r1, k: k1,b:b1}), fb.subs({r:r1, k: k1, b:b1}) ]).T
def DF(coord):
    global frr, frk, frb, fkr, fkk, fkb, fbr, fbk, fbb
    #el vector vertical tiene una dimensión más, es de la forma [ [r], [k]]
    r1 = coord[0,0]
    k1 = coord[1,0]
    b1 = coord[2,0]
    
    return np.matrix([
        [frr.subs({r:r1, k: k1, b:b1 }), frk.subs({r:r1, k: k1,b:b1}), frb.subs({r:r1, k: k1, b:b1})],
        [fkr.subs({r:r1, k: k1, b:b1 }), fkk.subs({r:r1, k: k1,b:b1}), fkb.subs({r:r1, k: k1, b:b1})],
        [fbr.subs({r:r1, k: k1, b:b1 }), fbk.subs({r:r1, k: k1,b:b1}), fbb.subs({r:r1, k: k1, b:b1})]
    ]).astype(float)


def Broyden(punto_inicial, useDF = False):
    print("punto inicial")
    print(punto_inicial)
    
    #---- Método de Newton de varias variables (Broyden) ------
    Nmax = 80; #numero máximo de iteraciones del método
    tol = 10**(-4); #tolerancia relativa del método
    pt=np.matrix(punto_inicial).T; # punto inicial, converge con (5,1)
    for i in range(0,Nmax):
        if i == 0:
            if useDF:
                df = DF(pt)
                Ainv = np.linalg.inv(df)
            else:
                Ainv = np.matrix('1 0 0; 0 1 0; 0 0 1')
        else:
            yk = F(pt1) - Fx
            sk=pt1-pt
            prod_ext = (sk-Ainv.dot(yk) ).dot(sk.T)
            numerador = prod_ext.dot(Ainv)
            denominador = sk.T.dot(Ainv).dot(yk)
            Ainv = Ainv + (1/denominador[0,0])*numerador
            pt=pt1;    
        Fx = F(pt)
        pt1= pt-Ainv.dot(Fx)
        if np.linalg.norm(pt1.astype(float) - pt.astype(float))/np.linalg.norm(pt1.astype(float)) < tol and i > 5:
            print("Tolerancia relativa alcanzada en ", str(i+1), " iterciones" )
            break 
        if i == Nmax-1:
            print("Número máximo de iteraciones alcanzado")
    pt = pt1
    #----------- Datos Obtenidos ---------------
    #print("Candidato a Raiz")
    #print(pt)
    #print("Evaluacion en f1 y f2");
    #print(F(pt,fr,fk))
    print("r, k y b obtenidos en escala")
    r = pt[0,0]
    k = pt[1,0]
    b = pt[2,0]
    print(r)
    print(k)
    print(b)
    #print("r y k reales")
    #print(r/max_ti)
    #print(k*max_p)

    #ya obtenidos r y k, volvemos a la escala real
    #ti = ti*max_ti;
    #p = p*max_p;

    #obtenemos P
    t0 = ti[0]
    p0 = p[0]
    P = funcion_poblacional(t0,p0,k,r,t,b)
    print("Evaluación en la función error")
    error = 0
    for i in range(0,n):
        x = ti[i];
        error = error + (P.subs(t,x) - p[i] )**2;
    print(error)
    #------------------------Gráfica ------------------------
    fig, ax = plt.subplots() #inicializamos grafico
    #intervalo a graficar
    desde_x = min(ti) 
    hasta_x = max(ti)
    #graficamos puntos dados
    ax.scatter(ti, p)
    #Graficamos la funcion
    xs = np.linspace(desde_x, hasta_x, 50) #intervalo y cantidad de puntos a evaluar
    ys = [P.subs(t,k) for k in xs] #evaluaciones en la funcion encontrada
    ax.plot(xs, ys, 'r-', label="P(t)") #graficamos con una recta
    #Estilizado de gráfica
    plt.xlabel("Tiempo") #etiqueta para eje x
    plt.ylabel("Población") #etiqueta para eje y
    plt.legend(loc='best') #habilitar la muestra de etiquetas
    plt.show()
    #limpiar gráfico para la siguiente iteración
    plt.figure().clear()
    plt.close()
    plt.cla()
    plt.clf()

########## ALGORITMO PRINCIPAL ###########
    
# Datos del problema
ti = np.arange(0,datos.shape[0],1) # x = 0,1,2,..., (numero de filas-1).
#ti = np.array(range(0,datos.shape[0]))
p = np.array(datos["M. Totales"])          
n = ti.shape[0] #cantidad de datos

#escalar la información, así no se desbordan las operaciones
max_ti = max(ti)
max_p = max(p)

ti= ti/max_ti
p= p/max_p

t0 = ti[0]
p0 = p[0] #primer dato poblacional (escalado)
#declaracion de símbols
t, r, k, b  = symbols('t r k b', real=True)
#funcion poblacional con variables simbólicas
P = funcion_poblacional(t0,p0,k,r,t,b)
#funcion error
E = sum( (P.subs(t,ti[i])-p[i])**2 for i in range(0,n))
print("Calculando derivadas parciales para armar F")
#derivadas parciales (sistema F)
fr = diff(E,r)
fk = diff(E,k)
fb = diff(E,b)
print("Calculando segundas derivadas parciales para armar DF")
#segundas derivadas parciales (matriz DF)
frr = diff(fr,r)
frk = diff(fr,k)
frb = diff(fr,b)
fkr = diff(fk,r)
fkk = diff(fk,k)
fkb = diff(fk,b)
fbr = diff(fb,r)
fbk = diff(fb,k)
fbb = diff(fb,b)    

#puntos iniciales    
rs = np.arange(310, 331,15)
ks = np.arange(0.9, 1.21, 0.2)
bs = np.arange(0.015, 0.036, 0.015)
puntos_iniciales = list(product(rs,ks,bs))
#puntos_iniciales = [(330, 1.0, 0.03), (330, 1.2000000000000002, 0.015), (330, 1.2000000000000002, 0.03)]
print("Iniciamos método de Broyden")
for punto_inicial in puntos_iniciales:
    try:
        Broyden(punto_inicial, useDF = True)
    except TypeError:
        print("Cannot convert complex to float. Siguiente punto...")
    #except LinAlgError:
    #    print("Matriz inicial DF no invertible. Siguiente punto...")
```

## Confirmar que los puntos hallados son mínimos


Teorema 9.6(CALCULUS volumen 2. Tom M.Apostol): (https://es.wikipedia.org/wiki/Matriz_hessiana)

Sea $f$ un campo escalar con derivadas parciales segundas continuas en una bola } $B(\textbf{a})$, designemos como $H(\textbf{a})$ a la matriz hessiana en el punto estacionario $\textbf{a}$. Entonces:

$f$ tiene un mínimo relativo en $\textbf{a}$ si todos los autovalores $H(\textbf{a})$ son positivos ssi los determinantes de las submatrices principales de $H(\textbf{a})$ son positivos (teorema de Análisis 1, 2 de Marzo del 2022)

Es decir, calcularemos los determinantes de las submatrices principales. Si todos son positivos, tenemos un mínimo relativo.

```python
def funcion_poblacional(t0,p0,k,r,t,b):
    #devuelve la función a usar uando los tipos y valores actuales de las variables
    #logistica (b=1): (k*p0)/(p0+(k-p0)*exp(-r*t))
    #r(P) cuadratica (b=2): P = p0*k/(  p0**2+(k**2-p0**2)*exp(-2*r*t) )**(1/2)
    #etc...
    #richards: p0*k/(  p0**b+(k**b-p0**b)*exp(-b*r*t) )**(1/b)
    return (k**b + (p0**b - k**b)*exp(-b*(r / k**b)*(t-t0)))**(1 / b)
# Datos del problema
ti = np.arange(0,datos.shape[0],1) # x = 0,1,2,..., (numero de filas-1).
#ti = np.array(range(0,datos.shape[0]))
p = np.array(datos["M. Totales"])          
n = ti.shape[0] #cantidad de datos

#escalar la información, así no se desbordan las operaciones
max_ti = max(ti)
max_p = max(p)

ti= ti/max_ti
p= p/max_p

t0 = ti[0]
p0 = p[0] #primer dato poblacional (escalado)
#declaracion de símbols
t, r, k, b  = symbols('t r k b', real=True)
#funcion poblacional con variables simbólicas
P = funcion_poblacional(t0,p0,k,r,t,b)
#funcion error
E = sum( (P.subs(t,ti[i])-p[i])**2 for i in range(0,n))
print("Calculando derivadas parciales para armar F")
#derivadas parciales (sistema F)
fr = diff(E,r)
fk = diff(E,k)
fb = diff(E,b)
print("Calculando segundas derivadas parciales para armar H")
#segundas derivadas parciales (matriz DF)
frr = diff(fr,r)
frk = diff(fr,k)
frb = diff(fr,b)
fkr = diff(fk,r)
fkk = diff(fk,k)
fkb = diff(fk,b)
fbr = diff(fb,r)
fbk = diff(fb,k)
fbb = diff(fb,b)

#puntos a evaluar
candidatos = [
    (237.757576784430,0.954066978461542,0.0353445352874852),
    (419.573183800770,0.950650410038468,0.0206151099295608)
    ]
    
for candidato in candidatos:
    print ("Punto")
    print(candidato)
    r1 = candidato[0]
    k1 = candidato[1]
    b1 = candidato[2]
    H = np.matrix([
        [frr.subs({r:r1, k: k1, b:b1 }), frk.subs({r:r1, k: k1,b:b1}), frb.subs({r:r1, k: k1, b:b1})],
        [fkr.subs({r:r1, k: k1, b:b1 }), fkk.subs({r:r1, k: k1,b:b1}), fkb.subs({r:r1, k: k1, b:b1})],
        [fbr.subs({r:r1, k: k1, b:b1 }), fbk.subs({r:r1, k: k1,b:b1}), fbb.subs({r:r1, k: k1, b:b1})]
    ]).astype(float)
    print("Determinantes de las submatrices principales")
    #submatriz 1
    ixgrid = np.ix_([True, True, True], [True, True, True])
    H1 = H[ixgrid]
    print(np.linalg.det(H1))
    #submatriz 2
    ixgrid = np.ix_([False, True, True], [False, True, True])
    H2 = H[ixgrid]
    print(np.linalg.det(H2))
    #submatriz 3
    ixgrid = np.ix_([False, False, True], [False, False, True])
    H3 = H[ixgrid]
    print(np.linalg.det(H3))
```

Debido a que no estamos exactamente en un punto mínimo local en cada caso, y que la computadora no hace cálculos exactos, basta con ver que los tres determinantes sean más "cercanos" a ser positivos o negativos. En todos los casos puede verse que los detemrinantes tienden a ser positivos, es decir, son puntos cercanos a sus respectivos **mínimos relativos**.
