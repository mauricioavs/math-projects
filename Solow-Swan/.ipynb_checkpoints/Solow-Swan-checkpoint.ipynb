# Estimación de parámetros de la ecuación diferencial fundamental de Solow-Swan

El propósito de este notebook es replicar los resultados de la siguiente publicación:

Brunner, N., Mayrpeter, G., & Kühleitner, M. (2022). Parameter estimation of the Solow–Swan fundamental differential equation. En Heliyon (Vol. 8, Issue 10, p. e10816). Elsevier BV. https://doi.org/10.1016/j.heliyon.2022.e10816

La ecuación diferencial con la que se tratará es 

$$ \frac{dk(t)}{dt}=p\cdot k(t)^a - q\cdot k(t), \hspace{2mm} k=K/L $$

La cual describe un modelo de acumulación de capital, en términos del capital ($K$)  y la fuerza laboral ($L$). Esta ecuación dispone de cuatro parámetros:

- exponente $a$
- parámetros escaladores $p$ y $q$
- el valor inicial $k(t_0)$, donde $t_0$ es el momento inicial

## Datos

Los datos a ajustar fueron obtenidos de www.ggdc.net/pwt de la tabla *Penn World Table* (obtenida el 6 de Febrero del 2023), el cual es un archivo que contiene parámetros económicos de diversos países desde 1950 hasta 2019.

```python
import pandas as pd
dataset = pd.read_csv("data.csv")
dataset
```

Para el capital, $K$, se uso de representante la columna $cn$ (capital stock in million US
dollars of 2017 at current purchasing power parities), y para la fuerza laboral $L$, se utilizó la columna $emp$ (number of persons engaged in millions), es decir, 
$$k=cn/emp$$

Los ajustes se hicieron para algunos países que tuvieran la información de estas columnas completa: Argentina, DR Congo, France, Japan, Pakistan, and USA.

Las fechas para las que se hizo cada ajuste fue de 1950 a 2019. Además, para USA, que se consideró un ajuste extra de 1970-2009 (decisión arbitraria para ver si el exponente $a$ se mantiene estable para los subperíodos).

El siguiente es un ejemplo de obtención de datos para France de 1950 a 2019:

```python
#import matplotlib.pyplot as plt
data = dataset[ dataset["country"] == "France"].copy()
data["k"] = data["cn"] / data["emp"]
data.plot(x = "year", y = "k", kind='scatter')
```

## Calibración

Para encontrar la solución no se usó la solución analítica de la ecuación (de hecho, se conoce y se muestra en el paper), sino que se realizó una implementación numérica.

Se usó el método de **mínimos cuadrados para ajustar los parámetros**, de forma que la solución $k(t)$ minimiza la suma de los errores cuadrados, $SSE$:

$$ SSE = \sum_{i=1}^N{(k_i-k(t_i))^2} $$

Con $N=70$ para la mayoría de los casos, $k_i$ el dato del momento $t_i$

Se usó también otra calibración $SSLE$, la cual es menos sensible a la **heterocedasticidad** (esto se da cuando la varianza de los errores no es constante en todas las observaciones realizadas):

$$ SSLE = \sum_{i=1}^N{(ln(k_i)-ln(k(t_i)))^2} $$

**Nota**: Los últimos tres puntos de France no se parecen a los mostrados en la gráfica correspondiente del paper, aunque los parámetros de ajuste dados si dan una buena curva como se muestra a continuación.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp #RK
dataset = pd.read_csv("data.csv")
data = dataset[ dataset["country"] == "France"].copy()
data["k"] = data["cn"] / data["emp"]
#-----Data------
ts = np.array(data["year"])
ys = np.array(data["k"])

t0 = min(ts) #initial time 
tf = max(ts) #final time 
def f(t,y,p, q, a):#definition of the vector field
    return p*y**a-q*y

# Runge-Kutta 5(4) solver -------------------------------
t_eval = ts
interval =[t0,tf]
k0, p, q, a = 9.87e4, 0.03, 3.08e-5, 0.99
sol=solve_ivp(f,interval, [k0], method='RK45',t_eval=t_eval,atol=1e-16,rtol=1e-8,
             args=(p, q, a)) #Runge-Kutta 45 method
error = np.sum(np.power(np.array(sol.y) - np.array(ys), 2))
print(error)
#Plotting the solution-------------------------------
fig, ax = plt.subplots() #inicializamos grafico
ax.scatter(ts, ys) #graficamos (x,y) dados del problema
ax.plot(sol.t, sol.y[0], "r-") #graficamos curva aproximada

plt.xlabel('t')
plt.ylabel('k(t)')
plt.show() 
```

## Primer implementación

Se usará el método de Runge-Kutta para obtener una aproximación de la curva a los datos de France con los parámetros dados, y  con ella y los puntos dados se obtendrá una aproximación de la función error $SSE$. 

Se optimizará el resultado usando diferencias finitas. Uno de los problemas que se enfrentan es conseguir una configuración inicial.

### ¿Cómo conseguir una configuración inicial?

Recordemos la ecuación diferencial a tratar.

$$ \frac{dk(t)}{dt}=p\cdot k(t)^a - q\cdot k(t) $$

Si observamos los puntos dados, parecería que una función esponencial sería una buen punto de partida por su forma de crecimiento. Para lograr esto, hagamos $q = 0 $ y $a=1$. De esta forma, la ecuación diferencial queda

$$ \frac{dk(t)}{dt}=p\cdot k(t) $$

De la cual conocemos la solución general

$$k(t)=c_1e^{pt} + c_2,\hspace{2mm} \text{con $c_1,c_2$ constantes}$$

Pero sabemos que la curva aproximada busca ser creciente, por lo que $c_1>0$, por lo que reescribimos la solución como

$$k(t)=e^{pt+c_1} + c_2,\hspace{2mm} \text{con $c_1,c_2$ constantes}$$

Falta encontrar $p$ y $k_0$.

Para encontrar $k_0$, debemos ajustarlo de tal forma que la curva aproximada pase cerca de los puntos. Como nuestra función es exponencial, resulta lógico ajustar $k_0$ como la ordenada del primer dato, pues después resta ajustar $p$ para mover la inclinación. 

**Nota:** Recordemos que $k_0$ es la ordenada donde Runge-kutta parte. En el caso de France, arbitrariamente se toma el primer punto $(1950, 83281.8)$ (obtenido de los datos "year" y "k") para pasar por él, es decir, 

$$k_0 = 83281.8 $$

Para encontrar $p$, debemos considerar que la aproximación por Runge-Kutta toma el valor actual de $k(t)$ para aproximar el siguiente (a un tamaño de paso), y debemos considerar $c_2 = 0$ (de otro modo Runge-Kutta no podría decidir cómo aproximar, pues habría infinitas exponenciales pasando por ese punto), es decir, 

$$k(1950) = e^{1950p+c_1} = 83281.8\implies z= 1950p+c_1 = \ln{83281.8}$$

Agregamos la condición de que la exponencial debe pasar por el punto final, este es (2019, 575778.034483), es decir,

$$k(1950) = e^{1950p+c_1} = e^z = 83281.8$$

$$k(2019) = e^{2019p+c_1}= e^{z+69p}  = 575778.034483$$

Si restamos estas ecuaciones implica que

$$e^{z+69p}-e^{z} = 492496.23448300007$$

La cual podemos resolver con un algoritmo de paso, pues conocemos $z$.

```python
p = 0.2 #start at this p
y = np.exp(np.log(83281.8) + 69*p)-np.exp(np.log(83281.8)) - 492496.23448300007 #objective is y(p) == 0
if (y < 0): #we initialize with some p that y(p) > 0
    print("use un p inicial más grande")
step = 0.5 #initial step
max_loops = 100
while  max_loops:
    p_new = p - step #new p
    y_new = np.exp(np.log(83281.8) + 69*p_new)-np.exp(np.log(83281.8)) - 492496.23448300007
    if y_new > 0:
        p = p_new
        y = y_new
    else:
        step/=2
    max_loops -= 1    
print("valor de p: " + str(p))
print("distancia de la curva al punto final: " + str(y))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp #RK
dataset = pd.read_csv("data.csv")
data = dataset[ dataset["country"] == "France"].copy()
data["k"] = data["cn"] / data["emp"]
#-----Data------
ts = np.array(data["year"])
ys = np.array(data["k"])

t0 = min(ts) #initial time 
tf = max(ts) #final time 
def f(t,y,p, q, a):#definition of the vector field
    return p*y**a-q*y

# Runge-Kutta 5(4) solver -------------------------------
t_eval = ts
interval =[t0,tf]
k0, p, q, a = 83281.8, 0.02802162596696435, 0, 1
sol=solve_ivp(f,interval, [k0], method='RK45',t_eval=t_eval,atol=1e-8,rtol=1e-8,
             args=(p, q, a)) #Runge-Kutta 45 method
#error = np.sum(np.power(np.array(sol.y) - np.array(ys), 2))
#print(error)
#Plotting the solution-------------------------------
fig, ax = plt.subplots() #inicializamos grafico
ax.scatter(ts, ys) #graficamos (x,y) dados del problema
ax.plot(sol.t, sol.y[0], "r-") #graficamos curva aproximada

plt.xlabel('t')
plt.ylabel('k(t)')
plt.show()
```

### Ajuste
Con la configuración inicial

$$k_0 = 83281.8$$
$$p = 0.02802162596696435$$
$$q=0$$
$$a = 1$$

podemos usar diferencias finitas para obtener una mejor aproximación.

```python
np.random.seed(1) #useful for replicating results
def get_gradient(curr_params):
    curr_pos = np.array(curr_params)
    curr_eval = main_func_eval(curr_pos)
    gradient = np.zeros(len(curr_pos))
    eps = 1.1e-16
    for i in range(len(curr_pos)):
        ei = np.zeros( len(curr_pos) )
        ei[i] = 1
        finite_diff = (main_func_eval( curr_params + eps*ei) - curr_eval)/eps
        gradient[i] = finite_diff
    return gradient

def main_func_eval(curr_params, graph = False):
    global interval, t_eval, ys 
    sol=solve_ivp(f,interval, [curr_params[0]], method='LSODA',t_eval=t_eval,atol=1e-08,rtol=1e-08,
             args=curr_params[1:]) #Runge-Kutta 45 method
    error = np.sum(np.power(np.array(sol.y) - np.array(ys), 2))
    if graph:
        fig, ax = plt.subplots() #inicializamos grafico
        ax.scatter(ts, ys) #graficamos (x,y) dados del problema
        ax.plot(sol.t, sol.y[0], "r-") #graficamos curva aproximada
        plt.xlabel('t')
        plt.ylabel('y(t)')
        plt.show() 
    return error

def f(t,y,p, q, a):#definition of the vector field
    return p*y**a-q*y
#-----Data------
dataset = pd.read_csv("data.csv")
data = dataset[ dataset["country"] == "France"].copy()
data["k"] = data["cn"] / data["emp"]

ts = np.array(data["year"])
ys = np.array(data["k"])

t0 = min(ts) #initial time 
tf = max(ts) #final time 
t_eval = ts
interval =[t0,tf]
#---------------
#prev_pos = [k0, p, q, a] form
prev_pos = np.array([83281.8, 0.02802162596696435, 0, 1])
prev_error = main_func_eval(prev_pos, graph = False)
step = 2
while step > 1e-16:
    gradient = get_gradient(prev_pos)
    gradient = gradient / np.linalg.norm(gradient) #unit length
    curr_pos = prev_pos - step * gradient
    curr_error = main_func_eval(curr_pos, graph = False)
    print(curr_error)
    if curr_error > prev_error or np.isnan(curr_error): #doesnt improve or new params dont make sense for eq.
        step/=2
        continue
    prev_pos = curr_pos
    prev_error = curr_error
best_error = main_func_eval(prev_pos, graph = True)
print("Minimum error")
print(best_error)
print("Best Params")
print("k0: ", str(prev_pos[0]))
print("p: ", str(prev_pos[1]))
print("q: ", str(prev_pos[2]))
print("a: ", str(prev_pos[3]))
```

Con los parámetros obtenidos podemos ver que, en términos del error, tenemos una mejor aproximación de la curva. No se debe olvidar que en el paper original utilizaron métricas estadísticas adicionales para decidir una buena aproximación, es decir, **un menor error no significa una mejor aproxomación general**.

```python
paper_error = 135098472845.90181 #obtanied from cell above
my_error = 129310652311.79573
my_error < paper_error
```

## Regresión lineal
https://www.cienciadedatos.net/documentos/py10-regresion-lineal-python.html

En los dos casos hemos visto que las soluciones son muy similares a una exponencial de la forma

$$y = \beta_0e^{\beta_1t}\epsilon$$

Y podemos linealizar esta expresion aplicando la función $\ln$:

$$\ln y = \ln\beta_0 +\beta_1 t + \ln \epsilon$$

De esta forma, podemos aplicar regresión lineal. Aplicamos esta función a los puntos de la base de datos.

```python
data = dataset[ dataset["country"] == "Mexico"].copy()
data["k"] = data["cn"] / data["emp"]
data["lnk"] = np.log(data["k"]) 
```

Cargamos los paquetes para realizar la regresión lineal.

```python
# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np

# Gráficos
# ==============================================================================
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns

# Preprocesado y modelado
# ==============================================================================
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Configuración matplotlib
# ==============================================================================
plt.rcParams['image.cmap'] = "bwr"
#plt.rcParams['figure.dpi'] = "100"
plt.rcParams['savefig.bbox'] = "tight"
style.use('ggplot') or plt.style.use('ggplot')

# Configuración warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

Graficamos las dos variables que queremos correlacionar.

```python
# Gráfico
# ==============================================================================
fig, ax = plt.subplots(figsize=(6, 3.84))

data.plot(
    x    = 'year',
    y    = 'lnk',
    c    = 'firebrick',
    kind = "scatter",
    ax   = ax
)
ax.set_title('Distribución de puntos');
```

Calculamos la correlación entre las dos variables

```python
# Correlación lineal entre las dos variables
# ==============================================================================
corr_test = pearsonr(x = data['year'], y =  data['lnk'])
print("Coeficiente de correlación de Pearson: ", corr_test[0])
print("P-value: ", corr_test[1])
```

El Coeficiente de correlación de Pearson cercano a uno indica que tenemos una correlación positiva fuerte, y el P-value muy pequeño indica que hay una correlación significativa entre las dos variables, por lo que resulta lógico relacionar estas variables con una regresión lineal.

### Ajuste
Hacemos el ajuste dividiendo datos para el entrenamiento y para la validación.

```python
# División de los datos en train y test
# ==============================================================================
X = data[['year']]
y = data['lnk']

X_train, X_test, y_train, y_test = train_test_split(
                                        X.values.reshape(-1,1),
                                        y.values.reshape(-1,1),
                                        train_size   = 0.8,
                                        random_state = 1234,
                                        shuffle      = True
                                    )
```

Hacemos el ajuste

```python
# A la matriz de predictores se le tiene que añadir una columna de 1s para el intercept del modelo
X_train = sm.add_constant(X_train, prepend=True)
modelo = sm.OLS(endog=y_train, exog=X_train,)
modelo = modelo.fit()
print(modelo.summary())
```

La columna *coef* contiene los coeficientes de nuestra regresión, es decir, el modelo lineal generado sigue la ecuación:

$$\ln k = -23.7605 + 0.0178t$$

El valor de *R-squared* indica que el modelo es capaz de explicar el 84.7% de la variabilidad observada en la variable respuesta (k).

Obtenemos intervalos de confianza para los coeficientes del modelo

```python
# Intervalos de confianza para los coeficientes del modelo
# ==============================================================================
modelo.conf_int(alpha=0.05)
```

Es decir, con una confianza del 95%, los valores de la regresión están en estos intervalos: 

$$\ln \beta_0 \in [-27.85, -19.68], \hspace{2mm} \beta_1 \in [0.015, 0.0198]$$

Hacemos las predicciones

```python
# Predicciones con intervalo de confianza del 95%
# ==============================================================================
predicciones = modelo.get_prediction(exog = X_train).summary_frame(alpha=0.05)
predicciones.head(4)
```

```python
# Predicciones con intervalo de confianza del 95%
# ==============================================================================
predicciones = modelo.get_prediction(exog = X_train).summary_frame(alpha=0.05)
predicciones['x'] = X_train[:, 1]
predicciones['y'] = y_train
predicciones = predicciones.sort_values('x')

# Gráfico del modelo
# ==============================================================================
fig, ax = plt.subplots(figsize=(6, 3.84))

ax.scatter(predicciones['x'], predicciones['y'], marker='o', color = "gray")
ax.plot(predicciones['x'], predicciones["mean"], linestyle='-', label="OLS")
ax.plot(predicciones['x'], predicciones["mean_ci_lower"], linestyle='--', color='red', label="95% CI")
ax.plot(predicciones['x'], predicciones["mean_ci_upper"], linestyle='--', color='red')
ax.fill_between(predicciones['x'], predicciones["mean_ci_lower"], predicciones["mean_ci_upper"], alpha=0.1)
ax.legend();
```

Error del test

```python
# Error de test del modelo 
# ==============================================================================
X_test = sm.add_constant(X_test, prepend=True)
predicciones = modelo.predict(exog = X_test)
rmse = mean_squared_error(
        y_true  = y_test,
        y_pred  = predicciones,
        squared = False
       )
print("")
print(f"El error (rmse) de test es: {rmse}")
```

El error de test del modelo es de 0.18. Las predicciones del modelo final se alejan en promedio 0.18 unidades del valor real.

Ahora predecimos los intervalos de valores futuros haciendo lo siguiente:

En la función que se modeló, $k(t)=cn/emp$, la variable cn representa el valor agregado bruto generado por el trabajo en un país en un año determinado, y emp representa el número total de horas trabajadas en ese mismo año.

Por lo tanto, si k(t) aumenta a lo largo del tiempo, esto puede indicar que el país está mejorando su productividad laboral, es decir, está generando más valor agregado por hora trabajada. Esto puede ser una señal de un mayor uso de tecnologías avanzadas, mejoras en la educación y capacitación de los trabajadores, entre otros factores que influyen en la productividad.

Por otro lado, si k(t) disminuye a lo largo del tiempo, esto puede ser una señal de que la productividad laboral del país está disminuyendo, lo que puede ser un problema para el crecimiento económico sostenible.

En general, la variable k(t) es una medida útil para evaluar la relación entre el valor agregado generado por el trabajo y el número total de horas trabajadas en un país a lo largo del tiempo, lo que puede ser útil para analizar los patrones de desarrollo económico y las políticas laborales de un país.

La variable k(t) = cn/emp se relaciona indirectamente con el PIB (Producto Interno Bruto) de un país, ya que la productividad laboral (representada por k(t)) es uno de los factores que influyen en la generación de la producción económica.

El PIB de un país se calcula como el valor de mercado de todos los bienes y servicios finales producidos dentro de un país en un período de tiempo determinado. La producción económica está influenciada por varios factores, incluyendo la cantidad y calidad del capital (k), la cantidad y calidad del trabajo (representada por emp), y la eficiencia con la que se utilizan estos factores para producir bienes y servicios (representada por k(t)).

Por lo tanto, si un país tiene una alta productividad laboral (es decir, k(t) es alto), es más probable que la producción económica sea mayor, lo que a su vez se refleja en un mayor PIB. Por otro lado, si la productividad laboral de un país es baja, es menos probable que la producción económica sea alta, lo que puede resultar en un PIB más bajo.

En resumen, aunque la variable k(t) no es directamente equivalente al PIB de un país, es un factor importante que influye en la producción económica y, por lo tanto, puede estar relacionada con el PIB. Sin embargo, para hacer una recomendación específica sobre el valor de crecimiento del PIB que se debe considerar en el presupuesto de la federación, **es necesario considerar otros factores económicos y sociales relevantes, como la situación actual del mercado laboral, la inflación, la balanza comercial, el nivel de inversión, entre otros.**
